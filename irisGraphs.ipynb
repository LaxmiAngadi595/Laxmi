```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.set()
plt.scatter(df.petal_length,df.sepal_width)
plt.ylabel('sepal_length')
plt.xlabel('petal_length')
plt.title('Iris_Length')
plt.show()
```

```python
plt.scatter(df.sepal_width,df.petal_width)
plt.ylabel('petal_width')
plt.xlabel('sepal_width')
plt.title('Iris_Width')
plt.show()
```

```python
sns.set()
plt.scatter(df.sepal_length,df.sepal_width,marker='+',color='r')
plt.scatter(df.petal_length,df.petal_width,marker='+',color='g')
plt.ylabel('sepal_width and petal_width')
plt.xlabel('sepal_length and petal_length')
plt.title('IRIS ')
plt.show()
```

```python
sns.set_style("darkgrid");
sns.FacetGrid(iris, hue="species", size=4) 
plt.scatter(iris.sepal_length, iris.sepal_width, color='r', marker='*')
plt.ylabel('sepal_width')
plt.xlabel('sepal_length')
plt.show();
```

```python
sns.set_style("darkgrid");
sns.FacetGrid(iris, hue="species", size=4) 
plt.scatter(iris.petal_length, iris.petal_width, color='g', marker='*')
plt.ylabel('petal_width')
plt.xlabel('petal_length')
plt.show();
```

```python
sns.set_style("darkgrid");
sns.FacetGrid(iris, hue="species", size=4) 
plt.scatter(iris.sepal_length, iris.petal_length, color='y', marker='d', edgecolor='k',alpha=0.7, sizes= (50,200), cmap='Blues')
plt.ylabel('petal_length')
plt.xlabel('sepal_length')
plt.show();
```

```python
sns.set()
plt.plot(iris.sepal_length,c='y',linewidth=1.5)
plt.plot(iris.sepal_width,c='k',linewidth=1.5)
plt.title('Iris-sepal')
plt.show()
```
